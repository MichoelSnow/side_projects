# Steps to build a high quality neural network 

1. Enable data augmentation, and precompute=True
1. Use `lr_find()` to find highest learning rate where loss is still clearly improving
1. Train last layer from precomputed activations for 1-2 epochs
1. Train last layer with data augmentation (i.e. precompute=False) for 2-3 epochs with cycle_len=1
1. Unfreeze all layers
1. Set earlier layers to 3x-10x lower learning rate than next higher layer
1. Use `lr_find()` again
1. Train full network with cycle_mult=2 until over-fitting

### Initial Settings

`sz` is the size that the images will be resized to in order to ensure that the training runs quickly.  Arch is the model architecture, which in this case is the <b>resnet34</b> model. resnet34 is a version of the model that won the 2015 ImageNet competition. Here is more info on [resnet models](https://github.com/KaimingHe/deep-residual-networks).

```python
sz=224
arch=resnet34
```

## 1. Enable data augmentation, and precompute=True

### Build the model

`ImageClassifierData` reads in the files and their labels using the specified class
1. from_paths: given as subfolder names
1. from_arrays: given as numpy arrays
1. from_csv: given as a CSV file

```python
data = ImageClassifierData.from_paths(PATH, tfms=tfms_from_model(arch, sz))
```

```python
zip(data.classes,data.val_y)
```

`ConvLearner` constructs but does not run a convolutional network

```python
learn = ConvLearner.pretrained(arch, data, precompute=True)
```

The `fit` function runs the actual neural network.  By default when we create a learner, it sets all but the last layer to frozen. That means that it's still only updating the weights in the last layer when we call fit.

```python
learn.fit(0.01, 2)
```

Get the predictions for the validation set (they are in log scale)

```python
log_preds = learn.predict()
log_preds.shape
```

## 2. Use lr_find() to find highest learning rate where loss is still clearly improving

It uses the technique developed in the 2015 paper [Cyclical Learning Rates for Training Neural Networks](http://arxiv.org/abs/1506.01186) where we simply keep increasing the learning rate from a very small value, until the loss starts decreasing. `lr_find()` may finish before going through each batch of examples if the loss decreases enough.

You will need to create a new learner first, because ??????

```python
learn = ConvLearner.pretrained(arch, data, precompute=True)
```

```python
lrf=learn.lr_find()
```

Visualize the lr as a functon of the iteration

```python
learn.sched.plot_lr()
```

Visualize the loss as a function of the learning rate.  A good rule of thumb for picking the learning rate is to find the minimum value and go back by one to two orders of magnitude.

```python
learn.sched.plot()
```

## 3. Train last layer from precomputed activations for 1-2 epochs

```python
learn.fit(1e-2, 1)
```

## 4. Train last layer with data augmentation (i.e. precompute=False) for 2-3 epochs with cycle_len=1

One way to avoid *overfitting* is to effectively create more data, through *data augmentation*. This refers to randomly changing the images in ways that shouldn't impact their interpretation, such as horizontal flipping, zooming, and rotating.  This is done by passing `aug_tfms` (*augmentation transforms*) to `tfms_from_model`, with a list of functions to apply that randomly change the image however we wish. For photos that are largely taken from the side (e.g. most photos of dogs and cats, as opposed to photos taken from the top down, such as satellite imagery) we can use the pre-defined list of functions `transforms_side_on`. We can also specify random zooming of images up to specified scale by adding the `max_zoom` parameter.

```python
tfms = tfms_from_model(resnet34, sz, aug_tfms=transforms_side_on, max_zoom=1.1)
```

Visualize the agumentations

```python
def get_augs():
    data = ImageClassifierData.from_paths(PATH, bs=2, tfms=tfms, num_workers=1)
    x,_ = next(iter(data.aug_dl))
    return data.trn_ds.denorm(x)[1]
```

```python
ims = np.stack([get_augs() for i in range(6)])
plots(ims, rows=2)
```

```python
data = ImageClassifierData.from_paths(PATH, tfms=tfms)
learn = ConvLearner.pretrained(arch, data, precompute=True)
```

```python
learn.fit(1e-2, 1)
```

```python
learn.precompute=False
learn.fit(1e-2, 3, cycle_len=1)
```
